## Importing the libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

## Getting  the data 

```python
from sklearn import datasets 
X,y= datasets.make_classification(n_samples = 50, n_features = 4,n_classes= 2,random_state=1)
df = pd.DataFrame(X, columns= ['f1','f2','f3','f4'])
df['target'] = y
```

```python
df.head()
```

```python
plt.style.use('seaborn-white')


plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(X, method='average'))
```

```python
# create a function for clustering accuracy 
def accuracy(actual, predicted):
    return np.sum(actual == predicted)/ len(actual)
```

# ALGGOMORATIVE CLUSTERING 

## Trial 1 :
### Using euclidean and average

```python
 # creating the model 
HCL1 = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='average')
# ‘average’ uses the average of the distances of each observation of the two sets.
HCL1.fit(X)
labels_1 = HCL1.labels_
# predicted labels 

# accuracy 
accuracy(y, labels_1)
```

## Visualization  : 1

```python
plt.style.use('seaborn-white')


fig,(axs1,axs2)= plt.subplots(1, 2,figsize=(10,5))

class_0 = df[y == 0]
class_1 = df[y == 1]
axs1.set_title('Actual label distribution')
axs1.scatter(class_0.f1, class_0.f2, color = 'blue')
axs1.scatter(class_1.f1, class_1.f2, color = 'red')



df['preds'] = labels_1
class_0_pred = df[labels_1 == 0]
class_1_pred = df[labels_1 == 1]


axs2.set_title('predicted label distribution')
axs2.scatter(class_0_pred.f1, class_0_pred.f2, color= 'blue')
axs2.scatter(class_1_pred.f1, class_1_pred.f2, color = 'red')
plt.show()
```

## Trial 2 : 
### using euclidean and ward

```python
 # creating the model 
HCL1 = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
# ‘ward’ minimizes the variance of the clusters being merged
HCL1.fit(X)
labels_2 = HCL1.labels_
# predicted labels 


# accuracy 
accuracy(y, labels_2)
```

## Visualization

```python
fig,(axs1,axs2)= plt.subplots(1, 2,figsize=(10,5))
class_0 = df[y == 0]
class_1 = df[y == 1]
axs1.set_title('Actual label distribution')
axs1.scatter(class_0.f1, class_0.f2, color = 'red')
axs1.scatter(class_1.f1, class_1.f2, color = 'blue')



df['preds'] = labels_1
class_0_pred = df[labels_2 == 0]
class_1_pred = df[labels_2 == 1]


axs2.set_title('predicted label distribution')
axs2.scatter(class_0_pred.f1, class_0_pred.f2, color = 'red')
axs2.scatter(class_1_pred.f1, class_1_pred.f2, color = 'blue')
plt.show()
```

## Trial 3: 
### using manhattan and average

```python
 # creating the model 
HCL1 = AgglomerativeClustering(n_clusters=2, affinity='manhattan', linkage='average')
HCL1.fit(X)
labels_3 = HCL1.labels_
# predicted labels 


# accuracy 
accuracy(y, labels_3)
```
